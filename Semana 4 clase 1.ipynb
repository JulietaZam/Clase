```python
df = pd.read_excel('datasets/mic.xlsx',
                   engine = 'openpyxl'
                  )
df
```

```python
counts = df ['Age'].value_counts()
counts 

type(counts)

counts= counts.to_frame()
type(counts)
counts
```

```python
counts.loc[:,'Age']
counts.loc[:,'Age'].sum(axis = 0)
counts['Total%']= (counts.loc[:,'Age'])/counts.loc[:,'Age'].sum(axis = 0)*100
```

```python
counts.reset_index(inplace=True)
counts.rename({"index":'Age_group'},
              axis = 1,
              inplace = True)
counts
```

```python
x = list(counts['Age_group'])
y = list(counts['Total%'])
mpl.style.use('ggplot') # grafica mas bonita JAJA
```

```python
plt.subplot(1,2,1)
plt.rcParams['figure.figsize']=(15,7)
plt.bar(x, y,
       width = 0.5, #ancho barra,
       color = 'darkcyan')
plt.xlabel('Age group', fontweight = 'regular', color = 'black',
          fontsize = 17, horizontalalignment = 'center')
plt.xticks(fontsize = 14)

plt.subplot(1,2,2)
plt.rcParams['figure.figsize']=(15,7)

counts['Total%'].plot(kind = 'pie',
                     )
```
