## Nueva grafica
si no sale grafica poner plt.show()

```python
x = np.arange(0, 3*np.pi, 0.1)
y = np.sin(x)

plt.plot(x,y)
```

## Otra grafica doble 
creo que sigue agarrando x pasada 

```python
ysin = np.sin(x)
ycos = np.cos(x)

plt.plot(x, ysin)
plt.plot(x, ycos)

plt.xlabel('X axis label')
plt.ylabel('Y axis label')

plt.title('Grafica')

plt.show()
```

## Muchas graficas
acomodar en orden 

```python
plt.subplot(2,2,1) # 2 renglones, 2 column, position 1 

plt.plot(x, ysin)
plt.title('Uno')

plt.subplot(2,2,2) # 2 renglones, 2 column, position 2
plt.plot(x, ycos)
plt.title('Dos')

plt.subplot(2,2,3) # 2 renglones, 2 column, position 1 
plt.plot(x, ysin)
plt.title('Tres')

plt.subplot(2,2,4) # 2 renglones, 2 column, position 2
plt.plot(x, ycos)
plt.title('Cuatro')

plt.show()
```

## Cargar base de datos
lo anterior fue para ver el nombre de la libreria

```python
path = 'datasets/Cartwheeldata.csv'
df = pd.read_csv(path)
df.head(2)
```

## Graficar algunas columnas de la base de datos
lmplot diagrama de dispersión 
fit_reg = False quitar linea de dispesion 
hue = 'Gender' separarlos en hombre y mujer 

```python
sns.lmplot(x = 'Wingspan',
           y = 'Height',
           data = df,
           fit_reg = False,
           hue = 'Gender',
          )
      
plt.title('Scatter plot of Wingspan vs Height')
plt.show
```

## Otro tipo de grafica
swarmplot

```python
sns.swarmplot(x = 'Gender', 
              y = 'Wingspan',
              data = df, 
             )
plt.title('Gender vs Wingspan, the Swarmplot')
plt.show
```
